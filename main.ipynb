```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customers.csv')
```

```python
country_counts = df['Country'].value_counts()

country_counts.plot(kind='bar', figsize=(10,6), color='green')
plt.title('Number of Customers by Country')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
top_cities = df['City'].value_counts().head(10)

top_cities.plot(kind='bar', figsize=(10,6), color='coral')
plt.title('Top 10 Cities by Number of Customers')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
country_counts = df['Country'].value_counts()
most_customers_country = country_counts.idxmax()  # Country with max customers
print("Country with most customers overall:", most_customers_country)
```
